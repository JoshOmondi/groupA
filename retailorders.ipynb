### Univariate analysis

```python
# used for anlysing of a single variable at a time
df.head()
```

```python
# histogram quantity distribution
sns.histplot(data= df, x= "Quantity")
plt.Figure(figsize= (10,7))
plt.title("Quantity Distribution")
plt.xlabel("Quantity of Goods")
plt.ylabel("Count")
```

```python
# histogram cost price distribution
sns.histplot(data= df, x= "cost price", bins= 4)
plt.Figure(figsize= (8,6))
plt.title("Cost Price")
plt.xlabel("price")
plt.ylabel("count")
```

```python
# box plot
# boxplot quantity distribution
sns.boxplot(data= df, x= "Quantity")
plt.Figure(figsize= (10,7))
plt.title("Quantity Distribution")
plt.xlabel("Quantity of Goods")
plt.ylabel("Count")
```

### Bivariate analysis

```python
#categorical vs categorical visualization
# bar plot(states and category)
total_states = df["State"].nunique()
print(total_states)

state_category= pd.crosstab(df["State"], df["Category"])
state_category.plot(kind= "bar")
plt.title("Category Orders State")
plt.xlabel("State")
plt.ylabel("Orders")
plt.xticks(rotation= 90)
plt.tight_layout
plt.show()
```

```python


state_category= pd.crosstab(df["Segment"], df["Category"])
state_category.plot(kind= "bar")
plt.title("Category Orders Segment")
plt.xlabel("Segment")
plt.ylabel("Count")
plt.xticks(rotation= 0)
plt.tight_layout
plt.show()
```
